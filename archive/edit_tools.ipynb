```python
import ipywidgets as widgets
from IPython.display import display, FileLink
from bs4 import BeautifulSoup

def convert_photos_in_html(input_html_path, output_html_path, photo_table):
    # Read the HTML file
    with open(input_html_path, 'r') as file:
        html_content = file.read()

    # Parse the HTML content
    soup = BeautifulSoup(html_content, 'html.parser')

    # Iterate over each photo entry in the table and replace divs with img tags
    for photo_class, photo_file in photo_table.items():
        # Find all divs with the corresponding photo class
        divs = soup.find_all("div", class_=photo_class)
        for div in divs:
            # Extract alt text if available
            figure = div.find("figure")
            inner_div = figure.find("div", role="img") if figure else None
            alt_text = inner_div.get('aria-label', '') if inner_div else ''
            
            # Create new img tag
            new_img_tag = soup.new_tag("img", src=photo_file, alt=alt_text)
            
            # Replace div with img tag
            div.replace_with(new_img_tag)

    # Save the modified HTML content to a new file
    with open(output_html_path, 'w') as file:
        file.write(str(soup))

    return output_html_path

# Photo conversion table
photo_conversion_table = {
    'photo1': 'IMG_4839 (2).jpg',
    'photo2': 'IMG_4840 (2).jpg',
    'photo3': 'IMG_4838 (2).jpg',
    'photo4': 'unnamed (2).jpg',
    'photo5': 'IMG-4899 (1).JPG',
    'photo6': 'IMG_4926.jpg',
    'photo7': 'IMG_2713.JPG',
    'photo8': 'IMG_5233 (2).jpg',
    'photo9': 'IMG_5232 (2).jpg',
    'photo10': 'IMG_5234 (2).jpg ',
    'photo11': 'IMG_5236 (2).jpg',
    'photo12': 'IMG_2701.JPG',
    'photo13': 'IMG_2697 (1).JPG',
    'photo14': 'IMG_2632 (4).JPG',
    'photo15': 'unnamed (3).jpg',
    'photo16': 'IMG_2677.JPG',
    'photo17': 'IMG_2719.JPG',
    'photo18': 'IMG_2659 (1).JPG',
    'photo19': 'D91CDA31-7765-4CB9-82F4-198924AA3BAD (1).JPG',
    'photo20': 'FullSizeRender (1).jpg',
    'photo21': 'IMG_2770 (3).JPG',
    'photo22': 'IMG_2753.JPG',
    'photo23': 'IMG_2807.JPG',
    'photo24': 'F3117EEF-A4CE-4B42-8FE3-2268B0CA2B16.JPG',
    'photo25': 'IMG_2786.JPG',
    'photo26': 'IMG_2826 (2).JPG',
    'photo27': 'IMG_2818 (1).JPG',
    'photo28': 'IMG_2842.JPG',
    'photo29': '0ymkzeb4ho221.jpg',
    'photo30': 'IMG_2725 (3).JPG',
    'photo31': 'IMG_2813 (2).JPG',
    'photo32': 'IMG_2862.JPG',
    'photo33': 'FullSizeRender (2).jpg',
    'photo34': 'IMG_2875 (2).JPG',
    'photo35': 'IMG_2855 (1).JPG',
    'photo36': 'IMG_2898.JPG',
    'photo37': 'IMG_2882.JPG',
    'photo38': 'IMG_2892.JPG',
    'photo39': 'IMG_2886.JPG',
    'photo40': 'Screen Shot 2017-10-27 at 4.30.07 PM.png',
    'photo41': 'IMG_2908.JPG',
    'photo42': 'IMG_2950.JPG',
    'photo43': 'IMG_2945.JPG',
    'photo44': 'IMG_3034 (3).JPG',
    'photo45': 'IMG_2966.JPG',
    'photo46': 'IMG_2989.JPG',
    'photo47': 'IMG_2915.JPG',
    'photo48': 'IMG_3131.JPG',
    'photo49': 'cover your cofee.JPG',
    'photo50': 'IMG_3416.JPG',
    'photo51': 'IMG_3441.JPG',
    'photo52': 'agra culture.JPG',
    'photo53': 'valley.JPG',
    'photo54': 'IMG_3585.JPG',
    'photo55': 'applebees4.JPG',
    'photo56': 'fab ferns1.JPG',
    'photo57': 'carbones2.JPG',
    'photo58': 'Chanhassen.jpg',
    'photo59': 'caremelos1.JPG'
}

# File upload widget
uploader = widgets.FileUpload(
    accept='.html',  # Accept only .html files
    multiple=False  # Only one file can be uploaded
)
display(uploader)

def on_upload_change(change):
    if uploader.value:
        input_file = next(iter(uploader.value))  # Get the uploaded file name
        content = uploader.value[input_file]['content']
        with open(input_file, 'wb') as f:  # Save the uploaded file locally
            f.write(content)
        
        output_html_path = 'modified_' + input_file
        convert_photos_in_html(input_file, output_html_path, photo_conversion_table)
        
        # Display a link to download the modified file
        display(FileLink(output_html_path))

uploader.observe(on_upload_change, names='value')
```
