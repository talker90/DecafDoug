```python
import re

def convert_placeholders_to_html(input_md_path, output_md_path):
    # Read the Markdown file
    with open(input_md_path, 'r') as file:
        markdown_content = file.read()

    # Regular expression pattern to match the placeholder syntax
    placeholder_pattern = re.compile(r'~img src="(.*?)"\n~alt="(.*?)"')

    # Function to replace placeholders with HTML format
    def replace_with_html(match):
        img_src = match.group(1)
        alt_text = match.group(2)
        return (f'<div style="text-align: center;">\n'
                f'    <img src="{img_src}" alt="{alt_text}" '
                'style="width: 50%; display: block; margin-left: auto; margin-right: auto;">\n'
                '</div>')

    # Replace all placeholders with HTML format
    converted_content = placeholder_pattern.sub(replace_with_html, markdown_content)

    # Save the modified content to a new Markdown file
    with open(output_md_path, 'w') as file:
        file.write(converted_content)

# Paths to the input and output Markdown files
input_md_path = 'index_1.md'
output_md_path = 'index_2.md'

# Run the conversion
convert_placeholders_to_html(input_md_path, output_md_path)

# Output path to download the modified Markdown file
print(f"Markdown file saved: {output_md_path}")
```
